```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma

def plot_gammas(params, x_min, x_max, log_x=False, num_points=1000):
    if log_x:
        x = np.logspace(np.log10(x_min), np.log10(x_max), num_points)
    else:
        x = np.linspace(x_min, x_max, num_points)

    plt.figure(figsize=(8, 5))

    for i, (alpha, scale) in enumerate(params):
        y = gamma.pdf(x, a=alpha, scale=scale)
        plt.plot(x, y, label=f"α={alpha:.2f}, θ={scale:.2g}")

    if log_x:
        plt.xscale("log")

    plt.xlabel("x")
    plt.ylabel("density")
    plt.title("Learnt Lengthscale Gamma Hyperpriors Per Dim")
    plt.legend()
    plt.grid(True, which="both", alpha=0.3)
    plt.tight_layout()
    plt.show()
```
